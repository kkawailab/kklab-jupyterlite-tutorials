## 1. matplotlib とは何か

**matplotlib** は、Python で最も広く使われているグラフ描画ライブラリです。

### matplotlib の特徴
- 折れ線グラフ、棒グラフ、散布図、ヒストグラムなど多様なグラフを作成可能
- グラフの細かい部分まで自由にカスタマイズできる
- pandas や NumPy と連携しやすい
- 論文や発表資料に使える高品質な図を出力できる

### pyplot とは
`matplotlib.pyplot` は、MATLAB に似たインターフェースを提供するモジュールです。
通常は `plt` という名前でインポートして使います。

## 3. グラフの装飾（タイトル、ラベル、凡例）

グラフを見やすくするためには、タイトルや軸ラベルが重要です。

### 3.1 タイトルと軸ラベル

````python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("サイン波のグラフ")    # タイトル
plt.xlabel("x（ラジアン）")      # x軸ラベル
plt.ylabel("sin(x)")             # y軸ラベル
plt.show()
````

### 3.2 凡例（legend）

````python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), label="sin(x)")  # label で凡例の名前を指定
plt.plot(x, np.cos(x), label="cos(x)")

plt.title("三角関数")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()  # 凡例を表示
plt.show()
````

### 3.3 凡例の位置

````python
x = np.linspace(0, 2 * np.pi, 100)

plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")

plt.title("凡例の位置を変更")
plt.legend(loc="lower left")  # 左下に配置
# loc の選択肢: 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'center' など
plt.show()
````

### 3.4 グリッド線

````python
x = np.linspace(0, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.title("グリッド線付きのグラフ")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)  # グリッド線を表示
plt.show()
````

## 4. 線のスタイルと色

グラフの線は、色、太さ、スタイル（実線、破線など）をカスタマイズできます。

### 4.1 色の指定

````python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), color='red')           # 色名で指定
plt.plot(x, np.sin(x + 1), color='#00FF00')   # 16進数で指定
plt.plot(x, np.sin(x + 2), color='b')         # 省略形（b=blue）

plt.title("色の指定方法")
plt.show()
````

### 主な色の指定方法

| 指定方法 | 例 |
|---------|----|
| 色名 | `'red'`, `'blue'`, `'green'`, `'orange'`, `'purple'` |
| 省略形 | `'r'`, `'b'`, `'g'`, `'c'`(cyan), `'m'`(magenta), `'y'`(yellow), `'k'`(black), `'w'`(white) |
| 16進数 | `'#FF0000'`, `'#00FF00'`, `'#0000FF'` |
| RGB タプル | `(1.0, 0.0, 0.0)` (0~1の範囲) |

### 4.2 線のスタイル

````python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), linestyle='-', label='実線 (solid)')      # 実線
plt.plot(x, np.sin(x) + 1, linestyle='--', label='破線 (dashed)')  # 破線
plt.plot(x, np.sin(x) + 2, linestyle='-.', label='一点鎖線 (dashdot)')  # 一点鎖線
plt.plot(x, np.sin(x) + 3, linestyle=':', label='点線 (dotted)')   # 点線

plt.title("線のスタイル")
plt.legend()
plt.show()
````

### 4.3 線の太さとマーカー

````python
x = np.linspace(0, 10, 20)

plt.plot(x, np.sin(x), linewidth=2, marker='o', label='丸マーカー')
plt.plot(x, np.cos(x), linewidth=1, marker='s', label='四角マーカー')

plt.title("線の太さとマーカー")
plt.legend()
plt.show()
````

### 主なマーカーの種類

| マーカー | 形状 |
|---------|------|
| `'o'` | 丸 |
| `'s'` | 四角 |
| `'^'` | 上三角 |
| `'v'` | 下三角 |
| `'*'` | 星 |
| `'+'` | プラス |
| `'x'` | バツ |
| `'D'` | ダイヤモンド |

### 4.4 フォーマット文字列（簡略記法）

````python
x = np.linspace(0, 10, 20)

# 'ro-' = 赤(r) + 丸マーカー(o) + 実線(-)
plt.plot(x, np.sin(x), 'ro-', label='赤・丸・実線')

# 'bs--' = 青(b) + 四角マーカー(s) + 破線(--)
plt.plot(x, np.cos(x), 'bs--', label='青・四角・破線')

# 'g^:' = 緑(g) + 三角マーカー(^) + 点線(:)
plt.plot(x, np.sin(x) + 0.5, 'g^:', label='緑・三角・点線')

plt.title("フォーマット文字列")
plt.legend()
plt.show()
````

## 5. 散布図（scatter plot）

**散布図** は、2つの変数の関係を点で表すグラフです。
相関関係の確認によく使われます。

### 5.1 基本の散布図

````python
np.random.seed(42)
x = np.random.randn(50)
y = x + np.random.randn(50) * 0.5  # x と正の相関

plt.scatter(x, y)
plt.title("基本の散布図")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

### 5.2 点の大きさと色

````python
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.rand(50)      # 色（0〜1の値）
sizes = np.random.rand(50) * 200  # 点のサイズ

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(label='色の値')
plt.title("大きさと色を変えた散布図")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

### 5.3 グループ別の散布図

````python
np.random.seed(42)

# グループA
x_a = np.random.randn(30) + 2
y_a = np.random.randn(30) + 2

# グループB
x_b = np.random.randn(30) - 1
y_b = np.random.randn(30) - 1

plt.scatter(x_a, y_a, c='red', label='グループA', alpha=0.7)
plt.scatter(x_b, y_b, c='blue', label='グループB', alpha=0.7)
plt.title("グループ別の散布図")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````

## 6. 棒グラフ（bar chart）

**棒グラフ** は、カテゴリごとの数値を比較するのに適したグラフです。

### 6.1 基本の棒グラフ

````python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 32, 48]

plt.bar(categories, values)
plt.title("基本の棒グラフ")
plt.xlabel("カテゴリ")
plt.ylabel("値")
plt.show()
````

### 6.2 横向きの棒グラフ

````python
categories = ['りんご', 'バナナ', 'みかん', 'ぶどう', 'いちご']
values = [150, 80, 120, 200, 180]

plt.barh(categories, values, color='coral')
plt.title("果物の価格")
plt.xlabel("価格（円）")
plt.show()
````

### 6.3 棒の色とエッジ

````python
categories = ['1月', '2月', '3月', '4月']
values = [100, 120, 90, 150]
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']

plt.bar(categories, values, color=colors, edgecolor='black', linewidth=1.5)
plt.title("月別売上")
plt.ylabel("売上（万円）")
plt.show()
````

### 6.4 グループ化された棒グラフ

````python
categories = ['1Q', '2Q', '3Q', '4Q']
product_a = [20, 25, 30, 35]
product_b = [15, 30, 25, 40]

x = np.arange(len(categories))  # カテゴリの位置
width = 0.35  # 棒の幅

plt.bar(x - width/2, product_a, width, label='製品A')
plt.bar(x + width/2, product_b, width, label='製品B')

plt.title("四半期別売上比較")
plt.xlabel("四半期")
plt.ylabel("売上")
plt.xticks(x, categories)  # x軸のラベルを設定
plt.legend()
plt.show()
````

### 6.5 積み上げ棒グラフ

````python
categories = ['1Q', '2Q', '3Q', '4Q']
product_a = [20, 25, 30, 35]
product_b = [15, 30, 25, 40]

plt.bar(categories, product_a, label='製品A')
plt.bar(categories, product_b, bottom=product_a, label='製品B')  # bottom で積み上げ

plt.title("四半期別売上（積み上げ）")
plt.xlabel("四半期")
plt.ylabel("売上")
plt.legend()
plt.show()
````

## 7. ヒストグラム（histogram）

**ヒストグラム** は、データの分布を可視化するためのグラフです。

### 7.1 基本のヒストグラム

````python
np.random.seed(42)
data = np.random.randn(1000)  # 標準正規分布

plt.hist(data, bins=30)
plt.title("正規分布のヒストグラム")
plt.xlabel("値")
plt.ylabel("頻度")
plt.show()
````

### 7.2 ビン数の影響

````python
np.random.seed(42)
data = np.random.randn(1000)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].hist(data, bins=10)
axes[0].set_title('bins=10')

axes[1].hist(data, bins=30)
axes[1].set_title('bins=30')

axes[2].hist(data, bins=50)
axes[2].set_title('bins=50')

plt.tight_layout()
plt.show()
````

### 7.3 複数のヒストグラムを重ねる

````python
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)    # 平均0、標準偏差1
data2 = np.random.normal(2, 1.5, 1000)  # 平均2、標準偏差1.5

plt.hist(data1, bins=30, alpha=0.5, label='グループ1')
plt.hist(data2, bins=30, alpha=0.5, label='グループ2')
plt.title("2つの分布の比較")
plt.xlabel("値")
plt.ylabel("頻度")
plt.legend()
plt.show()
````

### 7.4 密度ヒストグラム

````python
np.random.seed(42)
data = np.random.randn(1000)

plt.hist(data, bins=30, density=True, alpha=0.7, label='ヒストグラム')

# 理論的な正規分布曲線を重ねる
x = np.linspace(-4, 4, 100)
pdf = 1 / np.sqrt(2 * np.pi) * np.exp(-x**2 / 2)
plt.plot(x, pdf, 'r-', linewidth=2, label='正規分布')

plt.title("密度ヒストグラムと確率密度関数")
plt.xlabel("値")
plt.ylabel("密度")
plt.legend()
plt.show()
````

## 8. 円グラフ（pie chart）

**円グラフ** は、全体に対する各部分の割合を表すのに適したグラフです。

### 8.1 基本の円グラフ

````python
labels = ['りんご', 'バナナ', 'みかん', 'ぶどう']
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels)
plt.title("果物の販売比率")
plt.show()
````

### 8.2 パーセント表示と強調

````python
labels = ['A部門', 'B部門', 'C部門', 'D部門']
sizes = [35, 30, 20, 15]
explode = [0.1, 0, 0, 0]  # A部門を強調（離す）
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', 
        explode=explode, colors=colors, shadow=True)
plt.title("部門別売上比率")
plt.show()
````

### 8.3 ドーナツグラフ

````python
labels = ['製品A', '製品B', '製品C', 'その他']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', 
        wedgeprops=dict(width=0.5))  # width でドーナツ状に
plt.title("売上構成（ドーナツグラフ）")
plt.show()
````

## 9. グラフの保存

作成したグラフをファイルとして保存する方法を学びます。

### 9.1 savefig でファイル保存

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("保存サンプル")
plt.xlabel("x")
plt.ylabel("sin(x)")

# PNG形式で保存
plt.savefig('my_graph.png')
plt.show()

print("my_graph.png として保存しました")
````

### 9.2 解像度とサイズの指定

````python
x = np.linspace(0, 10, 100)
y = np.cos(x)

plt.figure(figsize=(10, 6))  # サイズを指定（インチ）
plt.plot(x, y)
plt.title("高解像度グラフ")

# 高解像度で保存（dpi = dots per inch）
plt.savefig('high_res_graph.png', dpi=300, bbox_inches='tight')
plt.show()

print("high_res_graph.png として保存しました（300 dpi）")
````

### 9.3 さまざまな形式での保存

```python
# PNG（ラスター形式、Web向け）
plt.savefig('graph.png')

# PDF（ベクター形式、論文向け）
plt.savefig('graph.pdf')

# SVG（ベクター形式、Web向け）
plt.savefig('graph.svg')

# JPEG（写真向け、グラフには不向き）
plt.savefig('graph.jpg')
```
