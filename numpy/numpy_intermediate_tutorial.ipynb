## 1. ç·šå½¢ä»£æ•°æ¼”ç®—

NumPy ã® `linalg` ãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ã¯ç·šå½¢ä»£æ•°ã®åŸºæœ¬æ¼”ç®—ã‚’æä¾›ã—ã¾ã™ã€‚

### 1.1 è¡Œåˆ—ã®ç©

````python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print('è¡Œåˆ— A:')
print(A)
print('\nè¡Œåˆ— B:')
print(B)

# è¦ç´ ã”ã¨ã®ç©ï¼ˆã‚¢ãƒ€ãƒãƒ¼ãƒ«ç©ï¼‰
print('\nè¦ç´ ã”ã¨ã®ç© (A * B):')
print(A * B)

# è¡Œåˆ—ã®ç©
print('\nè¡Œåˆ—ã®ç© (A @ B):')
print(A @ B)

# np.dot ã‚‚ä½¿ãˆã‚‹
print('\nnp.dot(A, B):')
print(np.dot(A, B))
````

### 1.2 ãƒ™ã‚¯ãƒˆãƒ«ã®å†…ç©ãƒ»å¤–ç©

````python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# å†…ç©
print(f'v1 = {v1}')
print(f'v2 = {v2}')
print(f'å†…ç© (v1 Â· v2): {np.dot(v1, v2)}')
print(f'å†…ç© (v1 @ v2): {v1 @ v2}')

# å¤–ç©ï¼ˆ3æ¬¡å…ƒãƒ™ã‚¯ãƒˆãƒ«ï¼‰
print(f'å¤–ç© (v1 Ã— v2): {np.cross(v1, v2)}')

# å¤–ç©ï¼ˆãƒ†ãƒ³ã‚½ãƒ«ç©ï¼‰
print('\nå¤–ç©ï¼ˆãƒ†ãƒ³ã‚½ãƒ«ç©ï¼‰:')
print(np.outer(v1, v2))
````

### 1.3 è¡Œåˆ—å¼ã¨é€†è¡Œåˆ—

````python
A = np.array([[1, 2],
              [3, 4]])

print('è¡Œåˆ— A:')
print(A)

# è¡Œåˆ—å¼
det = np.linalg.det(A)
print(f'\nè¡Œåˆ—å¼: {det:.4f}')

# é€†è¡Œåˆ—
A_inv = np.linalg.inv(A)
print('\né€†è¡Œåˆ—:')
print(A_inv)

# æ¤œè¨¼: A @ A_inv = I
print('\nA @ A_inv (å˜ä½è¡Œåˆ—ã«ãªã‚‹ã¯ãš):')
print(np.round(A @ A_inv, 10))
````

### 1.4 å›ºæœ‰å€¤ã¨å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«

````python
A = np.array([[4, 2],
              [1, 3]])

print('è¡Œåˆ— A:')
print(A)

# å›ºæœ‰å€¤ã¨å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f'\nå›ºæœ‰å€¤: {eigenvalues}')
print('å›ºæœ‰ãƒ™ã‚¯ãƒˆãƒ«ï¼ˆåˆ—ã”ã¨ï¼‰:')
print(eigenvectors)

# æ¤œè¨¼: A @ v = Î» @ v
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    print(f'\nå›ºæœ‰å€¤ {lam:.4f} ã®æ¤œè¨¼:')
    print(f'  A @ v = {A @ v}')
    print(f'  Î» * v = {lam * v}')
````

### 1.5 é€£ç«‹æ–¹ç¨‹å¼ã®è§£æ³•

````python
# é€£ç«‹æ–¹ç¨‹å¼: Ax = b
# 2x + y = 5
# x + 3y = 7

A = np.array([[2, 1],
              [1, 3]])
b = np.array([5, 7])

# è§£ã‚’æ±‚ã‚ã‚‹
x = np.linalg.solve(A, b)
print('é€£ç«‹æ–¹ç¨‹å¼ Ax = b ã®è§£:')
print(f'x = {x}')

# æ¤œè¨¼
print(f'\næ¤œè¨¼ (A @ x): {A @ x}')
print(f'b: {b}')
````

### 1.6 ç‰¹ç•°å€¤åˆ†è§£ï¼ˆSVDï¼‰

````python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print('è¡Œåˆ— A:')
print(A)

# ç‰¹ç•°å€¤åˆ†è§£
U, s, Vt = np.linalg.svd(A)

print('\nU (å·¦ç‰¹ç•°ãƒ™ã‚¯ãƒˆãƒ«):')
print(U)
print(f'\nç‰¹ç•°å€¤: {s}')
print('\nVt (å³ç‰¹ç•°ãƒ™ã‚¯ãƒˆãƒ«ã®è»¢ç½®):')
print(Vt)

# å¾©å…ƒ: A = U @ S @ Vt
S = np.zeros_like(A, dtype=float)
S[:len(s), :len(s)] = np.diag(s)
A_reconstructed = U @ S @ Vt
print('\nå¾©å…ƒã•ã‚ŒãŸè¡Œåˆ—:')
print(np.round(A_reconstructed, 10))
````

## 6. ãƒ™ã‚¯ãƒˆãƒ«åŒ–ã¨ãƒ‘ãƒ•ã‚©ãƒ¼ãƒãƒ³ã‚¹

### 6.1 ãƒ«ãƒ¼ãƒ— vs ãƒ™ã‚¯ãƒˆãƒ«åŒ–

````python
import time

# ã‚µãƒ³ãƒ—ãƒ«ãƒ‡ãƒ¼ã‚¿
np.random.seed(42)
data = np.random.randn(100000)

# ãƒ«ãƒ¼ãƒ—ã«ã‚ˆã‚‹è¨ˆç®—
def loop_sum_squares(arr):
    result = 0
    for x in arr:
        result += x ** 2
    return result

start = time.time()
result_loop = loop_sum_squares(data)
time_loop = time.time() - start

# ãƒ™ã‚¯ãƒˆãƒ«åŒ–ã«ã‚ˆã‚‹è¨ˆç®—
start = time.time()
result_vectorized = np.sum(data ** 2)
time_vectorized = time.time() - start

print(f'ãƒ«ãƒ¼ãƒ—: {time_loop:.6f} ç§’')
print(f'ãƒ™ã‚¯ãƒˆãƒ«åŒ–: {time_vectorized:.6f} ç§’')
print(f'é«˜é€ŸåŒ–ç‡: {time_loop / time_vectorized:.1f}å€')
````

### 6.2 np.vectorize

````python
# ã‚¹ã‚«ãƒ©ãƒ¼é–¢æ•°ã‚’é…åˆ—å¯¾å¿œã«å¤‰æ›
def scalar_func(x):
    if x < 0:
        return 0
    elif x < 1:
        return x
    else:
        return 1

# vectorize ã§ãƒ©ãƒƒãƒ—
vectorized_func = np.vectorize(scalar_func)

arr = np.array([-0.5, 0.3, 0.7, 1.2, 2.0])
result = vectorized_func(arr)
print(f'å…¥åŠ›: {arr}')
print(f'å‡ºåŠ›: {result}')
````

### 6.3 np.einsumï¼ˆã‚¢ã‚¤ãƒ³ã‚·ãƒ¥ã‚¿ã‚¤ãƒ³ç¸®ç´„è¨˜æ³•ï¼‰

````python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# è¡Œåˆ—ã®ç©
print('è¡Œåˆ—ã®ç© (ij,jk->ik):')
print(np.einsum('ij,jk->ik', A, B))

# è¦ç´ ã”ã¨ã®ç©
print('\nè¦ç´ ã”ã¨ã®ç© (ij,ij->ij):')
print(np.einsum('ij,ij->ij', A, B))

# ãƒˆãƒ¬ãƒ¼ã‚¹ï¼ˆå¯¾è§’æˆåˆ†ã®å’Œï¼‰
print(f'\nãƒˆãƒ¬ãƒ¼ã‚¹ (ii->): {np.einsum("ii->", A)}')

# è»¢ç½®
print('\nè»¢ç½® (ij->ji):')
print(np.einsum('ij->ji', A))
````
