## SciPy（問題 25-27）の解答

````python
# 問題 25 の解答
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

# 1. describe
result = stats.describe(data)
print("=== stats.describe() ===")
print(f"データ数: {result.nobs}")
print(f"平均: {result.mean:.2f}")
print(f"分散: {result.variance:.2f}")

# 2. 歪度と尖度
print(f"\n歪度: {stats.skew(data):.4f}")
print(f"尖度: {stats.kurtosis(data):.4f}")

# 3. 標準誤差
print(f"\n標準誤差 (SEM): {stats.sem(data):.4f}")
````

````python
# 問題 26 の解答
mu, sigma = 170, 6
dist = stats.norm(loc=mu, scale=sigma)

# 1. 180cm以上の確率
prob_180 = 1 - dist.cdf(180)
print(f"180cm以上の確率: {prob_180:.4f} ({prob_180*100:.2f}%)")

# 2. 160cm以下の確率
prob_160 = dist.cdf(160)
print(f"160cm以下の確率: {prob_160:.4f} ({prob_160*100:.2f}%)")

# 3. 90パーセンタイル
height_90 = dist.ppf(0.90)
print(f"上位10%に入る身長: {height_90:.2f}cm")
````

````python
# 問題 27 の解答
np.random.seed(42)

# 1. 乱数生成
data = stats.norm.rvs(loc=50, scale=10, size=200)

# 2. フィッティング
fitted_mu, fitted_sigma = stats.norm.fit(data)
print(f"推定されたパラメータ: μ={fitted_mu:.2f}, σ={fitted_sigma:.2f}")

# 3. 可視化
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, density=True, alpha=0.7, label='データ')

x = np.linspace(data.min(), data.max(), 100)
plt.plot(x, stats.norm.pdf(x, fitted_mu, fitted_sigma), 'r-', 
         linewidth=2, label=f'フィット (μ={fitted_mu:.1f}, σ={fitted_sigma:.1f})')

plt.title("ヒストグラムとフィットした正規分布")
plt.xlabel("値")
plt.ylabel("密度")
plt.legend()
plt.show()
````

## Scikit-learn（問題 28-30）の解答

````python
# 問題 28 の解答
from sklearn.preprocessing import StandardScaler, MinMaxScaler

X = np.array([[10, 1000],
              [20, 2000],
              [30, 3000],
              [40, 4000],
              [50, 5000]])

print("元のデータ:")
print(X)

# 1. 標準化
scaler_std = StandardScaler()
X_std = scaler_std.fit_transform(X)
print("\n標準化後:")
print(X_std)

# 2. Min-Max
scaler_mm = MinMaxScaler()
X_mm = scaler_mm.fit_transform(X)
print("\nMin-Maxスケーリング後:")
print(X_mm)
````

````python
# 問題 29 の解答
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

np.random.seed(42)
X = np.random.rand(50, 1) * 10
y = 3 * X.flatten() + 5 + np.random.randn(50) * 2

# 1. データ分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. モデル学習
model = LinearRegression()
model.fit(X_train, y_train)

# 3. 切片と係数
print(f"切片: {model.intercept_:.4f}")
print(f"係数: {model.coef_[0]:.4f}")

# 4. R²スコア
y_pred = model.predict(X_test)
print(f"R²スコア: {r2_score(y_test, y_pred):.4f}")

# 5. 可視化
plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.7, label='データ')
X_line = np.linspace(0, 10, 100).reshape(-1, 1)
plt.plot(X_line, model.predict(X_line), color='red', linewidth=2, label='回帰直線')
plt.title(f"線形回帰: y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
````

````python
# 問題 30 の解答
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# 1. データ読み込み
iris = load_iris()
X = iris.data
y = iris.target

# 2. データ分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. モデル学習
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# 4. 正解率
y_pred = model.predict(X_test)
print(f"正解率: {accuracy_score(y_test, y_pred):.4f}")

# 5. 混同行列
cm = confusion_matrix(y_test, y_pred)
print("\n混同行列:")
print(cm)

# 可視化
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('予測')
plt.ylabel('実際')
plt.title('混同行列')
plt.show()
````
