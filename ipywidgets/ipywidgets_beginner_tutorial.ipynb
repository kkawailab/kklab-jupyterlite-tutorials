## 4. ã‚¤ãƒ™ãƒ³ãƒˆå‡¦ç†ï¼ˆobserve, on_clickï¼‰

ã‚¦ã‚£ã‚¸ã‚§ãƒƒãƒˆã®å€¤ãŒå¤‰ã‚ã£ãŸã¨ãã‚„ã€ãƒœã‚¿ãƒ³ãŒã‚¯ãƒªãƒƒã‚¯ã•ã‚ŒãŸã¨ãã«å‡¦ç†ã‚’å®Ÿè¡Œã§ãã¾ã™ã€‚

### 4.1 observe: å€¤ã®å¤‰æ›´ã‚’ç›£è¦–

````python
# observe ã‚’ä½¿ã£ãŸå€¤ã®ç›£è¦–
slider_obs = widgets.IntSlider(value=50, min=0, max=100, description='å€¤:')
output_obs = widgets.Output()

def on_value_change(change):
    with output_obs:
        clear_output(wait=True)
        print(f"å¤ã„å€¤: {change['old']}")
        print(f"æ–°ã—ã„å€¤: {change['new']}")

slider_obs.observe(on_value_change, names='value')

display(slider_obs, output_obs)
````

````python
# ãƒ‰ãƒ­ãƒƒãƒ—ãƒ€ã‚¦ãƒ³ã®å¤‰æ›´ã‚’ç›£è¦–
color_dropdown = widgets.Dropdown(
    options=['èµ¤', 'é’', 'ç·‘', 'é»„'],
    value='èµ¤',
    description='è‰²:'
)
color_output = widgets.Output()

def on_color_change(change):
    color_map = {'èµ¤': 'red', 'é’': 'blue', 'ç·‘': 'green', 'é»„': 'yellow'}
    with color_output:
        clear_output(wait=True)
        color = color_map[change['new']]
        display(widgets.HTML(
            f"<div style='width:100px;height:50px;background:{color}'></div>"
        ))

color_dropdown.observe(on_color_change, names='value')

# åˆæœŸè¡¨ç¤º
on_color_change({'new': color_dropdown.value})

display(color_dropdown, color_output)
````

### 4.2 on_click: ãƒœã‚¿ãƒ³ã®ã‚¯ãƒªãƒƒã‚¯å‡¦ç†

````python
# ãƒœã‚¿ãƒ³ã‚¯ãƒªãƒƒã‚¯ã®ã‚¤ãƒ™ãƒ³ãƒˆå‡¦ç†
click_button = widgets.Button(description='ã‚¯ãƒªãƒƒã‚¯!', button_style='primary')
click_output = widgets.Output()
click_count = [0]  # ãƒªã‚¹ãƒˆã«ã—ã¦å‚ç…§ã‚’ä¿æŒ

def on_button_click(b):
    click_count[0] += 1
    with click_output:
        clear_output(wait=True)
        print(f"ã‚¯ãƒªãƒƒã‚¯å›æ•°: {click_count[0]} å›")

click_button.on_click(on_button_click)

display(click_button, click_output)
````

````python
# å˜ä½å¤‰æ›å™¨
conv_value = widgets.FloatText(value=1.0, description='å€¤:')
conv_from = widgets.Dropdown(
    options=['ãƒ¡ãƒ¼ãƒˆãƒ«', 'ã‚­ãƒ­ãƒ¡ãƒ¼ãƒˆãƒ«', 'ãƒã‚¤ãƒ«', 'ãƒ•ã‚£ãƒ¼ãƒˆ'],
    value='ãƒ¡ãƒ¼ãƒˆãƒ«',
    description='å¤‰æ›å…ƒ:'
)
conv_to = widgets.Dropdown(
    options=['ãƒ¡ãƒ¼ãƒˆãƒ«', 'ã‚­ãƒ­ãƒ¡ãƒ¼ãƒˆãƒ«', 'ãƒã‚¤ãƒ«', 'ãƒ•ã‚£ãƒ¼ãƒˆ'],
    value='ã‚­ãƒ­ãƒ¡ãƒ¼ãƒˆãƒ«',
    description='å¤‰æ›å…ˆ:'
)
conv_result = widgets.HTML(value='<p style="font-size:18px">çµæœ: </p>')

# ãƒ¡ãƒ¼ãƒˆãƒ«ã¸ã®å¤‰æ›ä¿‚æ•°
to_meter = {
    'ãƒ¡ãƒ¼ãƒˆãƒ«': 1,
    'ã‚­ãƒ­ãƒ¡ãƒ¼ãƒˆãƒ«': 1000,
    'ãƒã‚¤ãƒ«': 1609.34,
    'ãƒ•ã‚£ãƒ¼ãƒˆ': 0.3048
}

def convert(change=None):
    value = conv_value.value
    from_unit = conv_from.value
    to_unit = conv_to.value
    
    # ä¸€åº¦ãƒ¡ãƒ¼ãƒˆãƒ«ã«å¤‰æ›ã—ã¦ã‹ã‚‰ç›®çš„ã®å˜ä½ã«å¤‰æ›
    meters = value * to_meter[from_unit]
    result = meters / to_meter[to_unit]
    
    conv_result.value = f'<p style="font-size:18px"><b>{value} {from_unit}</b> = <b>{result:.4f} {to_unit}</b></p>'

# å„ã‚¦ã‚£ã‚¸ã‚§ãƒƒãƒˆã®å¤‰æ›´ã‚’ç›£è¦–
conv_value.observe(convert, names='value')
conv_from.observe(convert, names='value')
conv_to.observe(convert, names='value')

# åˆæœŸè¨ˆç®—
convert()

converter_ui = widgets.VBox([
    widgets.HTML('<h3>å˜ä½å¤‰æ›å™¨</h3>'),
    conv_value,
    conv_from,
    conv_to,
    conv_result
])

display(converter_ui)
````

### 8.2 ç°¡æ˜“é›»å“

````python
# ç°¡æ˜“é›»å“
calc_display = widgets.Text(
    value='0',
    disabled=True,
    layout=widgets.Layout(width='220px')
)

current_value = [0]
pending_operation = [None]
new_number = [True]

def make_button(text, style=''):
    btn = widgets.Button(
        description=text,
        layout=widgets.Layout(width='50px', height='40px'),
        button_style=style
    )
    return btn

def on_digit(b):
    if new_number[0]:
        calc_display.value = b.description
        new_number[0] = False
    else:
        if calc_display.value == '0':
            calc_display.value = b.description
        else:
            calc_display.value += b.description

def on_operation(b):
    if pending_operation[0] is not None:
        on_equals(None)
    current_value[0] = float(calc_display.value)
    pending_operation[0] = b.description
    new_number[0] = True

def on_equals(b):
    if pending_operation[0] is None:
        return
    second = float(calc_display.value)
    op = pending_operation[0]
    if op == '+':
        result = current_value[0] + second
    elif op == '-':
        result = current_value[0] - second
    elif op == 'Ã—':
        result = current_value[0] * second
    elif op == 'Ã·':
        result = current_value[0] / second if second != 0 else 'Error'
    
    if result != 'Error':
        calc_display.value = str(result) if result == int(result) else f'{result:.6g}'
    else:
        calc_display.value = 'Error'
    pending_operation[0] = None
    new_number[0] = True

def on_clear(b):
    calc_display.value = '0'
    current_value[0] = 0
    pending_operation[0] = None
    new_number[0] = True

# ãƒœã‚¿ãƒ³ä½œæˆ
digits = [make_button(str(i)) for i in range(10)]
for d in digits:
    d.on_click(on_digit)

ops = [make_button(op, 'info') for op in ['+', '-', 'Ã—', 'Ã·']]
for op in ops:
    op.on_click(on_operation)

equals_btn = make_button('=', 'success')
equals_btn.on_click(on_equals)

clear_btn = make_button('C', 'danger')
clear_btn.on_click(on_clear)

# ãƒ¬ã‚¤ã‚¢ã‚¦ãƒˆ
row1 = widgets.HBox([digits[7], digits[8], digits[9], ops[3]])  # 7 8 9 Ã·
row2 = widgets.HBox([digits[4], digits[5], digits[6], ops[2]])  # 4 5 6 Ã—
row3 = widgets.HBox([digits[1], digits[2], digits[3], ops[1]])  # 1 2 3 -
row4 = widgets.HBox([clear_btn, digits[0], equals_btn, ops[0]]) # C 0 = +

calculator = widgets.VBox([
    widgets.HTML('<h3>ç°¡æ˜“é›»å“</h3>'),
    calc_display,
    row1, row2, row3, row4
])

display(calculator)
````

### 8.3 ToDo ãƒªã‚¹ãƒˆ

````python
# ToDo ãƒªã‚¹ãƒˆ
todo_input = widgets.Text(placeholder='æ–°ã—ã„ã‚¿ã‚¹ã‚¯ã‚’å…¥åŠ›...', layout=widgets.Layout(width='250px'))
todo_add_btn = widgets.Button(description='è¿½åŠ ', button_style='primary')
todo_list_container = widgets.VBox()

todos = []

def update_list():
    items = []
    for i, (task, done) in enumerate(todos):
        checkbox = widgets.Checkbox(value=done, indent=False, layout=widgets.Layout(width='30px'))
        label_style = 'text-decoration: line-through; color: gray;' if done else ''
        label = widgets.HTML(f'<span style="{label_style}">{task}</span>')
        delete_btn = widgets.Button(description='Ã—', button_style='danger', layout=widgets.Layout(width='30px'))
        
        def make_toggle(idx):
            def toggle(change):
                todos[idx] = (todos[idx][0], change['new'])
                update_list()
            return toggle
        
        def make_delete(idx):
            def delete(b):
                todos.pop(idx)
                update_list()
            return delete
        
        checkbox.observe(make_toggle(i), names='value')
        delete_btn.on_click(make_delete(i))
        
        items.append(widgets.HBox([checkbox, label, delete_btn]))
    
    todo_list_container.children = items

def add_todo(b):
    task = todo_input.value.strip()
    if task:
        todos.append((task, False))
        todo_input.value = ''
        update_list()

todo_add_btn.on_click(add_todo)

# Enter ã‚­ãƒ¼ã§ã‚‚è¿½åŠ 
def on_enter(change):
    if change['new']:
        add_todo(None)

todo_ui = widgets.VBox([
    widgets.HTML('<h3>ToDo ãƒªã‚¹ãƒˆ</h3>'),
    widgets.HBox([todo_input, todo_add_btn]),
    todo_list_container
])

display(todo_ui)
````

### 8.4 ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–ãªã‚°ãƒ©ãƒ•ï¼ˆMatplotlib é€£æºï¼‰

````python
# matplotlib ã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ï¼ˆå¿…è¦ãªå ´åˆï¼‰
try:
    import piplite
    await piplite.install(["matplotlib", "numpy"])
except ImportError:
    pass

import numpy as np
import matplotlib.pyplot as plt
````

````python
# ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–ãªã‚µã‚¤ãƒ³æ³¢ãƒ—ãƒ­ãƒƒãƒˆ
freq_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='å‘¨æ³¢æ•°:')
amp_slider = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='æŒ¯å¹…:')
phase_slider = widgets.FloatSlider(value=0.0, min=0.0, max=2*np.pi, step=0.1, description='ä½ç›¸:')
plot_output = widgets.Output()

def update_plot(change=None):
    with plot_output:
        clear_output(wait=True)
        
        freq = freq_slider.value
        amp = amp_slider.value
        phase = phase_slider.value
        
        x = np.linspace(0, 2*np.pi, 500)
        y = amp * np.sin(freq * x + phase)
        
        fig, ax = plt.subplots(figsize=(8, 4))
        ax.plot(x, y, 'b-', linewidth=2)
        ax.set_xlim(0, 2*np.pi)
        ax.set_ylim(-2.5, 2.5)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title(f'y = {amp:.1f} Ã— sin({freq:.1f}x + {phase:.1f})')
        ax.grid(True, alpha=0.3)
        ax.axhline(y=0, color='k', linewidth=0.5)
        plt.tight_layout()
        plt.show()

# å„ã‚¹ãƒ©ã‚¤ãƒ€ãƒ¼ã®å¤‰æ›´ã‚’ç›£è¦–
freq_slider.observe(update_plot, names='value')
amp_slider.observe(update_plot, names='value')
phase_slider.observe(update_plot, names='value')

# åˆæœŸãƒ—ãƒ­ãƒƒãƒˆ
update_plot()

plot_ui = widgets.VBox([
    widgets.HTML('<h3>ã‚¤ãƒ³ã‚¿ãƒ©ã‚¯ãƒ†ã‚£ãƒ–ãªã‚µã‚¤ãƒ³æ³¢</h3>'),
    freq_slider,
    amp_slider,
    phase_slider,
    plot_output
])

display(plot_ui)
````
